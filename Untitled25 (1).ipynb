**Q4: What is a Common Table Expression (CTE) in SQL, and what are its main
benefits? Provide a simple SQL example demonstrating its usage.**

**Ans:* A CTE (Common Table Expression) is a temporary result set that can be referenced within a SELECT,INSERT,UPDATE,orDELETE statement.

Benefits:

* Improves readability
* Makes complex queries easier to manage

Example:



```python
import sqlite3

# Connect to existing database (your ECommerceDB)
conn = sqlite3.connect("ECommerceDB.db")
cursor = conn.cursor()

# Create a sample Orders table if not already there
cursor.executescript("""
DROP TABLE IF EXISTS Orders;
CREATE TABLE Orders(
    OrderID INTEGER PRIMARY KEY,
    CustomerID INTEGER,
    TotalAmount REAL
);

INSERT INTO Orders VALUES
(1001, 1, 1245.50),
(1002, 2, 99.99),
(1003, 1, 145.00),
(1004, 3, 150.00),
(1005, 4, 120.00),
(1006, 1, 20.00);
""")

# ‚úÖ Run CTE query
query = """
WITH TotalOrders AS (
    SELECT
        CustomerID,
        SUM(TotalAmount) AS TotalSpent
    FROM Orders
    GROUP BY CustomerID
)
SELECT
    CustomerID,
    TotalSpent
FROM TotalOrders
WHERE TotalSpent > 500;
"""

print("Customers whose total order amount > 500:")
for row in cursor.execute(query):
    print(row)

conn.close()


```

**Q9: Write a SQL query that uses a Common Table Expression (CTE) and a
Window Function (specifically ROW_NUMBER() or RANK()) to display the
CategoryName, ProductName, and Price for the top 2 most expensive products in
each CategoryName.**

```python
import sqlite3
import pandas as pd

# 1Ô∏è‚É£ Database & Cursor setup
conn = sqlite3.connect(':memory:')
cur = conn.cursor()

# 2Ô∏è‚É£ Create Tables
cur.execute('''
CREATE TABLE customer (
    customer_id INTEGER PRIMARY KEY,
    first_name TEXT,
    last_name TEXT,
    email TEXT
)
''')

cur.execute('''
CREATE TABLE payment (
    payment_id INTEGER PRIMARY KEY,
    customer_id INTEGER,
    amount REAL,
    payment_date TEXT,
    FOREIGN KEY (customer_id) REFERENCES customer(customer_id)
)
''')

cur.execute('''
CREATE TABLE category (
    category_id INTEGER PRIMARY KEY,
    name TEXT
)
''')

cur.execute('''
CREATE TABLE film (
    film_id INTEGER PRIMARY KEY,
    title TEXT,
    category_id INTEGER,
    rental_rate REAL,
    FOREIGN KEY (category_id) REFERENCES category(category_id)
)
''')

cur.execute('''
CREATE TABLE inventory (
    inventory_id INTEGER PRIMARY KEY,
    film_id INTEGER,
    store_id INTEGER,
    FOREIGN KEY (film_id) REFERENCES film(film_id)
)
''')

cur.execute('''
CREATE TABLE rental (
    rental_id INTEGER PRIMARY KEY,
    customer_id INTEGER,
    inventory_id INTEGER,
    rental_date TEXT,
    FOREIGN KEY (customer_id) REFERENCES customer(customer_id),
    FOREIGN KEY (inventory_id) REFERENCES inventory(inventory_id)
)
''')

conn.commit()

# 3Ô∏è‚É£ Insert Sample Data
cur.executemany('INSERT INTO customer VALUES (?,?,?,?)', [
    (1,'Ekta','Patel','ekta@example.com'),
    (2,'Rahul','Singh','rahul@example.com'),
    (3,'Priya','Sharma','priya@example.com')
])

cur.executemany('INSERT INTO payment VALUES (?,?,?,?)', [
    (1,1,150.0,'2025-08-01'),
    (2,1,200.0,'2025-09-05'),
    (3,2,300.0,'2025-08-12')
])

cur.executemany('INSERT INTO category VALUES (?,?)', [
    (1,'Action'),
    (2,'Comedy')
])

cur.executemany('INSERT INTO film VALUES (?,?,?,?)', [
    (1,'Movie A',1,2.99),
    (2,'Movie B',1,4.99),
    (3,'Movie C',2,3.99),
    (4,'Movie D',2,5.49)
])

cur.executemany('INSERT INTO inventory VALUES (?,?,?)', [
    (1,1,1),
    (2,2,1),
    (3,3,2),
    (4,4,2)
])

cur.executemany('INSERT INTO rental VALUES (?,?,?,?)', [
    (1,1,1,'2025-08-01'),
    (2,1,2,'2025-08-02'),
    (3,2,3,'2025-09-05'),
    (4,3,4,'2025-09-10')
])

conn.commit()

# üßæ ------------- Q9 -------------
print("\nQ9: Top 2 Most Expensive Films per Category (CTE + Window Function)\n")
query9 = '''
WITH RankedFilms AS (
    SELECT
        c.name AS CategoryName,
        f.title AS ProductName,
        f.rental_rate AS Price,
        ROW_NUMBER() OVER (PARTITION BY c.name ORDER BY f.rental_rate DESC) AS rnk
    FROM film f
    JOIN category c ON f.category_id = c.category_id
)
SELECT CategoryName, ProductName, Price
FROM RankedFilms
WHERE rnk <= 2
ORDER BY CategoryName, Price DESC;
'''
display(pd.read_sql_query(query9, conn))
```

```python
import sqlite3
import pandas as pd

# 1Ô∏è‚É£ Database & Cursor setup
conn = sqlite3.connect(':memory:')
cur = conn.cursor()

# 2Ô∏è‚É£ Create Tables
cur.execute('''
CREATE TABLE customer (
    customer_id INTEGER PRIMARY KEY,
    first_name TEXT,
    last_name TEXT,
    email TEXT
)
''')

cur.execute('''
CREATE TABLE payment (
    payment_id INTEGER PRIMARY KEY,
    customer_id INTEGER,
    amount REAL,
    payment_date TEXT,
    FOREIGN KEY (customer_id) REFERENCES customer(customer_id)
)
''')

cur.execute('''
CREATE TABLE category (
    category_id INTEGER PRIMARY KEY,
    name TEXT
)
''')

cur.execute('''
CREATE TABLE film (
    film_id INTEGER PRIMARY KEY,
    title TEXT,
    category_id INTEGER,
    rental_rate REAL,
    FOREIGN KEY (category_id) REFERENCES category(category_id)
)
''')

cur.execute('''
CREATE TABLE inventory (
    inventory_id INTEGER PRIMARY KEY,
    film_id INTEGER,
    store_id INTEGER,
    FOREIGN KEY (film_id) REFERENCES film(film_id)
)
''')

cur.execute('''
CREATE TABLE rental (
    rental_id INTEGER PRIMARY KEY,
    customer_id INTEGER,
    inventory_id INTEGER,
    rental_date TEXT,
    FOREIGN KEY (customer_id) REFERENCES customer(customer_id),
    FOREIGN KEY (inventory_id) REFERENCES inventory(inventory_id)
)
''')

conn.commit()

# 3Ô∏è‚É£ Insert Sample Data
cur.executemany('INSERT INTO customer VALUES (?,?,?,?)', [
    (1,'Ekta','Patel','ekta@example.com'),
    (2,'Rahul','Singh','rahul@example.com'),
    (3,'Priya','Sharma','priya@example.com')
])

cur.executemany('INSERT INTO payment VALUES (?,?,?,?)', [
    (1,1,150.0,'2025-08-01'),
    (2,1,200.0,'2025-09-05'),
    (3,2,300.0,'2025-08-12')
])

cur.executemany('INSERT INTO category VALUES (?,?)', [
    (1,'Action'),
    (2,'Comedy')
])

cur.executemany('INSERT INTO film VALUES (?,?,?,?)', [
    (1,'Movie A',1,2.99),
    (2,'Movie B',1,4.99),
    (3,'Movie C',2,3.99),
    (4,'Movie D',2,5.49)
])

cur.executemany('INSERT INTO inventory VALUES (?,?,?)', [
    (1,1,1),
    (2,2,1),
    (3,3,2),
    (4,4,2)
])

cur.executemany('INSERT INTO rental VALUES (?,?,?,?)', [
    (1,1,1,'2025-08-01'),
    (2,1,2,'2025-08-02'),
    (3,2,3,'2025-09-05'),
    (4,3,4,'2025-09-10')
])

conn.commit()

# üßæ ------------- Q10 (a) Top 5 customers by spending -------------
print("\nQ10(a): Top 5 Customers by Total Spent\n")
query10a = '''
SELECT
    c.first_name || ' ' || c.last_name AS CustomerName,
    SUM(p.amount) AS TotalSpent
FROM customer c
JOIN payment p ON c.customer_id = p.customer_id
GROUP BY c.customer_id
ORDER BY TotalSpent DESC
LIMIT 5;
'''
display(pd.read_sql_query(query10a, conn))

# üßæ ------------- Q10 (b) Top 3 categories by rental count -------------
print("\nQ10(b): Top 3 Categories by Rental Count\n")
query10b = '''
SELECT
    c.name AS CategoryName,
    COUNT(r.rental_id) AS RentalCount
FROM rental r
JOIN inventory i ON r.inventory_id = i.inventory_id
JOIN film f ON i.film_id = f.film_id
JOIN category c ON f.category_id = c.category_id
GROUP BY c.name
ORDER BY RentalCount DESC
LIMIT 3;
'''
display(pd.read_sql_query(query10b, conn))
```
